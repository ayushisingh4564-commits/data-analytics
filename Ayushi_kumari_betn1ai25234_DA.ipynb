**Case Study 1: Student Marks Analysis**

  Problem Statement

Analyze student marks using NumPy to perform matrix operations and statistical analysis.

```python
import numpy as np

# Dataset: Marks of 5 students in 3 subjects
marks = np.array([
    [80, 75, 90],
    [60, 88, 70],
    [95, 92, 85],
    [70, 65, 60],
    [88, 78, 84]
])

# 1. Total marks of each student
total_marks = np.sum(marks, axis=1)

# 2. Average marks per student
average_marks = np.mean(marks, axis=1)

# 3. Identify topper
topper_index = np.argmax(total_marks)

# 4. Subject-wise highest score
subject_highest = np.max(marks, axis=0)

# 5. Standard deviation of each subject
subject_std = np.std(marks, axis=0)

print("Total Marks:", total_marks)
print("Average Marks:", average_marks)
print("Topper is Student:", topper_index + 1)
print("Subject-wise Highest Scores:", subject_highest)
print("Subject-wise Standard Deviation:", subject_std)
```

**Output Explanation & Interpretation**

1.The third student has the highest total marks, making them the topper.

2.Subject-wise highest scores indicate strong academic performance in       Subject 1 and Subject 2.

3.The standard deviation values show moderate variation among students.

4.Lower standard deviation in Subject 3 suggests more consistent performance compared to other subjects.

**Performance Consistency**

The dataset shows that Student 3 performs consistently well across all subjects. Subject 3 has relatively stable marks among students, indicating balanced performance levels.


**Case Study 3: Sales Analysis**

   Problem Statement

Analyze weekly sales data to understand growth trends and variability.

1.   List item
2.   List item



```python
import numpy as np

# Dataset: Weekly Sales
sales = np.array([1200, 1500, 1100, 1700, 1600, 1800, 1400])

# 1. Average daily sales
avg_sales = np.mean(sales)

# 2. Day with highest sales
highest_day = np.argmax(sales)

# 3. Percentage growth from Day 1 to Day 7
growth_percentage = ((sales[-1] - sales[0]) / sales[0]) * 100

# 4. Cumulative sales
cumulative_sales = np.cumsum(sales)

# 5. Standard deviation
sales_std = np.std(sales)

print("Average Daily Sales:", avg_sales)
print("Highest Sales on Day:", highest_day + 1)
print("Percentage Growth:", growth_percentage)
print("Cumulative Sales:", cumulative_sales)
print("Standard Deviation:", sales_std)
```

**Output Explanation & Interpretation**

The average daily sales provide an overall business performance measure.

Sales peak on Day 6, indicating strong demand that day.

Positive percentage growth from Day 1 to Day 7 shows business expansion.

The cumulative sales curve reflects steady revenue accumulation.

Standard deviation indicates moderate variability in daily sales performance.

  Sales Growth Pattern

The data shows an overall upward trend with slight fluctuations. The peak on Day 6 suggests possible promotional impact or increased customer demand.

**Conclusion Section**

 Insights Derived

Student analysis revealed performance differences and subject-wise consistency.

Sales analysis showed a positive growth trend with moderate daily fluctuations.

Standard deviation helped measure variability in both academic and business datasets.

Cumulative and percentage calculations provided trend-based insights.
